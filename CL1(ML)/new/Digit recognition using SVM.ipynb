# Implementation of Support Vector Machines (SVM) for classifying images of hand-written digits into their respective numerical classes (0 to 9).

## Loading dataset

```python
from sklearn.datasets import load_digits
```

```python
X = load_digits()['data']
y = load_digits()['target']
```

```python
X.shape,y.shape
```

# Spliting data into training and test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape
```

## Fitting SVM

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train,y_train)
```

```python
svc.score(X_test,y_test)
```

# Testing

```python
import matplotlib.pyplot as plt

i = 34
plt.imshow(X_train[i].reshape((8,8)));
plt.title(f"Actual: {y_train[i]} Predicted: {svc.predict([X_train[i]])[0]}");
```

# Classification Metric

```python
from sklearn.metrics import classification_report

y_preds = svc.predict(X_test)

print(classification_report(y_test,y_preds))
```
